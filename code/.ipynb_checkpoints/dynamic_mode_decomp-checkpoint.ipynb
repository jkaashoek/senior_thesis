### Section 2: Data Reading and Processing

```python
# Read in John's Hopkins data. We read in both case counts and deaths.

county_cases = pd.read_csv("../data/time_series_covid19_confirmed_US.csv")
county_deaths = pd.read_csv("../data/time_series_covid19_deaths_US.csv")

county_deaths.head()
```

```python
# Toggle to switch between using deaths and cases. This results presented in the paper are with county deaths
county_cases = county_deaths.copy()
```

```python
# Filter data to get to the 50 states + District of Columbia

# All provinces
provinces = list(county_cases.Province_State.unique())
# Provinces to remove
to_remove = ['Diamond Princess', 'Guam', 'Grand Princess', 'Northern Mariana Islands', 'Puerto Rico', 'Virgin Islands']
# States to keep 
states = [x for x in provinces if x not in to_remove]
# Filter
county_cases = county_cases[county_cases.Province_State.isin(states)]
county_cases.head()
```

```python
# Drop any rows that do not have a FIPS (unique county identifier) ID

county_cases = county_cases[~county_cases.FIPS.isna()].reset_index(drop=True)
county_cases.head()
```

```python
# Remove counties with no cases

# The names of columns with dates
dates = county_cases.columns[12:]
# The date range we will use throughout the code
date_range = dates[SMOOTH-1:]
# Number of days 
n_days = len(date_range)
# The number of days to perform our retrospective analysis with (this was toggled for some robustness checks)
m = n_days
# To be considered for the retrospective analysis, counties must have at least 5 deaths in this range (i.e. none of the
# rows in our X matrix for DMD can have less than 5 deaths)
zero_range = date_range[:m-1]

# Array to hold indices of counties with less than 5 deaths during zero_range
zero_cases = []

# Find all counties with less than 5 deaths
for i in range(len(county_cases)):
    if county_cases[zero_range].values[i].sum() < 5:
        zero_cases.append(i)
        
# Drop the counties with zero cases
county_cases = county_cases.drop(zero_cases, axis=0)
```

```python
# Get the number of new cases per day, casting anything that is less than 0 (or nan values) to 0

# These are the columns we want to keep on top of the date columns
keep_cols= ["FIPS", "Population", "Province_State", "Admin2"]

new_cases = county_cases[keep_cols+[dates[0]]].copy()
for i in range(1, len(dates)):
    curr_date = dates[i]
    last_date = dates[i-1]
    diff = county_cases[curr_date] - county_cases[last_date]
    new_cases[curr_date] = np.where((diff >= 0), diff, 0) 

new_cases.head()
```

```python
# Apply a smoothing by averaging over the last SMOOTH number of days

smoothed_df = county_cases[keep_cols].copy()
for i in range(SMOOTH-1, len(dates)):
    sm_range = dates[i-SMOOTH+1:i+1]
    smoothed_df[dates[i]] = np.mean(new_cases[sm_range], axis=1)

smoothed_df.head()            
```

```python
# The dataframe we will use for the rest of the analysis

df = smoothed_df.copy()
```

### Section 3: Retrospective Analysis and Figures

```python
# Set up and calculate SVD for DMD

# Data matrices
all_data = df[date_range].values
all_data = scipy.stats.zscore(all_data, axis=1)
X_orig = df[date_range].values[:,:m-1]
X_prime_orig = df[date_range].values[:,1:m]

# Convert to z-scores
X = scipy.stats.zscore(X_orig, axis=1)
X_prime = scipy.stats.zscore(X_prime_orig, axis=1)

# SVD
u,s,vh = np.linalg.svd(all_data, full_matrices=True)
v = np.copy(vh.T)

# Plot the singular values
plt.figure()
plt.plot(s)
plt.xlabel('svd mode')
plt.ylabel('value')
plt.title('singular values of X')
plt.show()

# Plot the variance explained by each singular value
total = sum(s)
explained_var = [s[i] / total for i in range(len(s))]
plt.figure()
plt.plot(explained_var, color=colors[4])
plt.xticks([i for i in range(len(s))][::40], [i for i in range(len(s))][::40], fontproperties=prop, fontsize=10)
yticks = plt.gca().get_yticks()
plt.yticks([round(i,2) for i in yticks[1:]], [round(i,2) for i in yticks[1:]], fontproperties=prop, fontsize=10)
plt.title('Fraction of total energy stored in singular values', fontproperties=prop, fontsize=12)
plt.xlabel('Index of singular value', fontproperties=prop, fontsize=11)
plt.ylabel('Fraction', fontproperties=prop, fontsize=11)
sns.despine(top=True, right=True, left=True, bottom=True)
plt.savefig('../figures/chapter_2/energy_plot.pdf', dpi=200, bbox_inches='tight')
plt.show()

# Plot the first three right singular vectors
plot_dates = [q for q in date_range if "/1/" in q]
date_idxs = [list(date_range).index(q) for q in plot_dates]
plt.figure()
plt.plot(v[:,0].real*-1, label='0')
plt.plot(v[:,1].real, label='1')
plt.plot(v[:,2].real*-1, label='2')
plt.xticks(date_idxs, plot_dates, rotation=30, fontproperties=prop, fontsize=10)
yticks = plt.gca().get_yticks()
plt.yticks([round(i,2) for i in yticks], [round(i,2) for i in yticks], fontproperties=prop, fontsize=10)
plt.title('First three columns of v', fontproperties=prop, fontsize=12)
plt.xlabel("Time", fontproperties=prop, fontsize=11)
plt.ylabel("Disease severity (z-score)", fontproperties=prop, fontsize=11)
prop.set_size(10)
plt.legend(prop=prop, frameon=False)
sns.despine(top=True, right=True,left=True, bottom=True)
plt.savefig("../figures/chapter_2/svd.pdf", dpi=200, bbox_inches='tight')
plt.show()
```

```python
## We can start in on DMD

# SVD X, a matrix of the first m-1 timesteps
u,s,vh = np.linalg.svd(X, full_matrices=False)
print("original svd shapes", u.shape, s.shape, vh.shape)
v = vh.T

# Number of columns to use in the reconstruciton
r = 180
ur, sr, vr = u[:,:r], np.diag(s[:r]), v[:,:r]
# Calculate A_tilde, the approximate propogator
A_tilde = ur.T@X_prime@vr@np.linalg.inv(sr)

# Get the eigenvalues and vectors of A_tilde
vals,vecs = np.linalg.eig(A_tilde)
# Calculate the dynamic modes
phi = X_prime@vr@np.linalg.inv(sr)@vecs
# Get our omega values
omega = np.log(vals)
```

```python
## Plot the eigenspectrum

# Function to change the shade of a color based on v, a float between 0 and 1
def f(w,c,v):
    return (c - w) * v + w

# Get the magnitude of each eigenvalue
mags = [abs(q) for q in vals]
# Scale to be between 0 and 1
mags = mags / np.max(np.abs(mags))
# Set the color of a given point based on the magntiude
cols = [f(255,128,q) for q in mags]
# Set of t values to plot a unit cirlce
t = np.linspace(0,np.pi*2,100)

# Get the real and complex parts of the eigenvluaes
r = [x.real for x in vals]
c = [x.imag for x in vals]

# Plot the eigenspectrum
plt.figure()
plt.scatter(r, c, c=cols)
plt.plot(np.cos(t), np.sin(t), linewidth=1, label='unit circle', color=colors[5])
xticks, yticks = plt.gca().get_xticks()[1:-1], plt.gca().get_yticks()[1:-1]
plt.xticks([round(i,2) for i in xticks], [round(i,2) for i in xticks], fontproperties=prop, fontsize=10)
plt.yticks([round(i,2) for i in yticks], [round(i,2) for i in yticks], fontproperties=prop, fontsize=10)
plt.xlabel('Real part', fontproperties=prop, fontsize=11)
plt.ylabel('Complex part', fontproperties=prop, fontsize=11)
plt.title('Eigenvalues of DMD', fontproperties=prop, fontsize=12)
plt.legend(prop=prop, frameon=False, bbox_to_anchor=(0.75,0.97))
sns.despine(top=True, bottom=True, left=True, right=True)
plt.gca().set_aspect('equal', 'box')
plt.savefig("../figures/chapter_2/dmd_eigenvals.pdf", dpi=200, bbox_inches='tight')
plt.show()
```

```python
## Construct the power plot (power vs. freq)

# Power to use in lambda_j^p ||phi[:,j] || calculation
p = 100

# Calculate the frequencies for each eigenvalue
freqs = np.array([np.log(i).imag / (2*np.pi) for i in vals])
# Calculate lambda_j^p ||phi[:,j] ||
ys = np.array([abs(vals[i])**p * np.linalg.norm(phi[:,i]) for i in range(len(vals))])
# Get indices of positive frequncies (we will only plot these because the negative frequencies 
# are the same, just reflected across the y axis)
idxs = np.where(freqs > 0)[0]
# Indices chosen for plotting
chosen_idxs = [78, 81, 83, 84, 85]

# Construct the plot
f, ax = plt.subplots() 
ax.scatter(freqs[idxs], ys[idxs], color=colors[7])


# This code labels each point based on its index. Uncomment this to add the index to the plot, which is how I chose
# which dynamic modes to plot
# for i in range(len(freqs[idxs])):
#     ax.annotate(str(i), (freqs[idxs][i], ys[idxs][i]))

for i in chosen_idxs:
    print(1/freqs[idxs][i])
    ax.scatter(freqs[idxs][i], ys[idxs][i], color=colors[6], label='selected mode' if i==chosen_idxs[0] else '')

xticks, yticks = ax.get_xticks()[1:-1], ax.get_yticks()[1:-1]
plt.xticks([round(i,2) for i in xticks], [round(i,2) for i in xticks], fontproperties=prop, fontsize=10)
plt.yticks([round(i,2) for i in yticks], [round(i,2) for i in yticks], fontproperties=prop, fontsize=10)
plt.xlabel("Frequency", fontproperties=prop, fontsize=11)
plt.ylabel("Power", fontproperties=prop, fontsize=11)
plt.title("Choosing Dynamic modes", fontproperties=prop, fontsize=12)
ax.legend(frameon=False)
sns.despine(top=True, bottom=True, left=True, right=True)
plt.savefig("../figures/chapter_2/power_plot.pdf", dpi=200)
plt.show()
```

```python
## Function to generate maps of the U.S.
from mpl_toolkits.axes_grid1 import make_axes_locatable, inset_locator

def get_county_colors(county, cmap):
    '''
    Get shade and transparency for a given county based on its phase value
    '''
    ph = county.phase
    if np.isnan(ph):
        c, alpha = "white", 1        
    else:
        c, alpha = cmap(ph), 1
    return c, alpha

def get_cmap(n_levels, phase=True):
    '''
    Get colormap dependent on if we are plotting phase or magnitude
    '''
    if phase:
        cmap = mpl.cm.get_cmap('twilight', n_levels+5)
    else:
        cmap = mpl.cm.get_cmap('inferno', n_levels+3)
    return cmap

def get_color_scale(vec, phase, n_levels=15):
    '''
    Rescale vector and establish a custom color scale based on quantiles so that 
    counties are colored based on relative magnitudes
    '''
    # Rescale vector
    small, large = np.nanmin(vec), np.nanmax(vec)
    vec = (vec - small) / (large-small)
    
    # Equal number of values in each bucket
    buckets = np.unique(pd.qcut(vec, q=n_levels).astype(str))
    buckets = buckets[buckets != 'nan']
    
    # Get endpoints of buckets and ensure 0 and 1 are included
    lims =  np.sort([float(q[1:q.find(',')]) for q in buckets])
    lims[0] = 0
    lims = np.append(lims,1)
    
    # Get the original values that correspond to bucket limits
    original_vals = [round(q, 4) for q in lims * (large-small) + small]
    
    # Create mapping
    cmap = get_cmap(n_levels, phase)
    colors = [str(mpl.colors.rgb2hex(cmap(q))) for q in range(phase*2, cmap.N-2)][::-1]
    mapping = list(zip(lims, colors))
    cmap = mpl.colors.LinearSegmentedColormap.from_list('custom map', mapping)
    return vec, cmap, original_vals, lims


def add_county_geometries(ax, counties, cmap):
    '''
    Add all counties in a given state to the map
    '''
    for i in range(len(counties)):
        county = counties.iloc[i]
        c, alpha = get_county_colors(county, cmap)
        ax.add_geometries([county.geometry], 
                          crs=ccrs.PlateCarree(), 
                          facecolor=c, # add color based on vec
                          edgecolor=c)
        
def add_colorbar(ax, cmap, lims, original_vals, rescale=False):
    '''
    Add a colorbar across the bottom of the map based on our custom mapping
    '''
    axins1 = inset_locator.inset_axes(ax,
                    width="80%",  
                    height="5%",  
                    bbox_to_anchor=(0, -0.21, 1, 1),
                    bbox_transform=ax.transAxes,
                    loc='lower left')
    sm = mpl.cm.ScalarMappable(cmap=cmap)
    cbar = f.colorbar(sm, cax=axins1, orientation='horizontal')
    labs = ['%.3f'%(q) for q in original_vals]
    print(labs)
    if rescale:
        cbar.set_ticks([lims[0], 1])
        cbar.set_ticklabels([labs[0], labs[-2]])
    else:
        cbar.set_ticks(lims)
        cbar.set_ticklabels(labs)
    cbar.outline.set_linewidth(0.5)
    for t in axins1.xaxis.majorTicks:
        t.label1.set_fontproperties(prop)
        t.label1.set_fontsize(8)

    
# A function that draws inset map, ++
# ===================================
## https://stackoverflow.com/questions/55598249/showing-alaska-and-hawaii-in-cartopy-map
def add_insetmap(axes_extent, map_extent, state_geo, state_counties, cmap, ax):
    '''
    Inset on map for Alaska and Hawaii
    '''
    # create new axes, set its projection
    use_projection = ccrs.Mercator()     # preserve shape well
    sub_ax = plt.axes(axes_extent, projection=use_projection)  # normal units
    sub_ax.set_extent(map_extent)  # map extents

    # Add counties
    add_county_geometries(sub_ax, state_counties, cmap)
    sub_ax.add_geometries([state_geo], facecolor='none',  edgecolor='black', lw=0.2, crs=ccrs.PlateCarree())
    sub_ax.axis('off')
    
def get_inset_pos(rescale, pos):
    '''
    Get locations on axes for Alaska and Hawaii based on the type of figure
    '''
    if rescale:
        axes_extent_ak = (pos.x0+0.001, pos.y0+0.02, 0.2*(pos.x1-pos.x0), 0.2*(pos.y1-pos.y0))
        axes_extent_hi = (pos.x0+0.018, pos.y0+0.02, 0.15*(pos.x1-pos.x0), 0.15*(pos.y1-pos.y0))
    else:
        axes_extent_ak = (pos.x0+0.015, pos.y0+0.060, 0.18*(pos.x1-pos.x0), 0.18*(pos.y1-pos.y0))
        axes_extent_hi = (pos.x0+0.038, pos.y0+0.059, 0.12*(pos.x1-pos.x0), 0.12*(pos.y1-pos.y0))
    return axes_extent_ak, axes_extent_hi
                          
def plot_vector_on_map(ax, vec, phase=True, rescale=True, colorbar=True):
    '''
    Function to plot a vector, vec, on a map of the United States using the given matplotlib axis
    '''
    plot_df = df.copy()
    if rescale:
        vec, cmap, original_vals, lims = get_color_scale(vec, phase)
    else:
        cmap = mpl.cm.get_cmap('bwr')
        interval = np.hstack([np.linspace(0, 0.3), np.linspace(0.7, 1)])
        cmap = mpl.colors.LinearSegmentedColormap.from_list('cut_map', cmap(interval))
        lims, original_vals = [0, 0.5, 1], [0, 0.5, 1]
        vec = (vec - MIN_V) / (MAX_V - MIN_V)
        
    plot_df['phase'] = vec
    
    # Merge onto shapefile of county shapes
    plot_county = county_shp.merge(plot_df, on='fips', how='left') 
    # Establish counties and states
    counties = list(county_shp.geometry.values)
    states = list(state_shp.geometry.values)
    
    pos = ax.get_position()
    ak, hi = get_inset_pos(rescale, pos)
    
    # Plout counties state by state
    for i in range(len(state_shp)):
        state = state_shp.iloc[i]
        state_num = state.STATEFP
        # Alaska inset
        if  state_num == '02':
            map_extent = (-178, -128, 49, 71)    
            add_insetmap(ak, map_extent, state.geometry, plot_county[plot_county.STATEFP == state_num], cmap, ax)
        # Hawaii inset
        elif state_num == '15':
            map_extent = (-159, -154, 18.5, 22)
            add_insetmap(hi, map_extent, state.geometry, plot_county[plot_county.STATEFP == state_num], cmap, ax)
        # Contiguous states
        else:
            add_county_geometries(ax, plot_county[plot_county.STATEFP == state_num], cmap)
            ax.add_geometries([state.geometry], facecolor='none', edgecolor='black', lw=0.2, alpha=0.5, crs=ccrs.PlateCarree())
    
    # Add colorbar to bottom of map
    if colorbar:
        add_colorbar(ax, cmap, lims, original_vals, rescale=rescale)  
    
    # Set limits on what we want to show in the map (these values limit our plot to just mainland U.S.)
    ax.set_extent([-125, -66.8, 22, 48])
    # Turn off the axes for a cleaner plot
    ax.axis('off')
    
    return
```

```python
## Generate visualization of DMD modes - this takes a minute or two to run

def gen_tseries(value, time_stamps):
    '''
    Function to calaulcate e^(omega*t), where t is determined by time_stamps
    '''
    return np.exp(value*time_stamps)
    
# Number of modes to plot
n_modes = len(original_idxs)
# Vector of time stamps
t = np.linspace(0, len(date_range), len(date_range)+1)

# Plot 
f = plt.figure(figsize=(15,10))
# Set the ratios of each column
spec = gridspec.GridSpec(nrows=n_modes, ncols=3, figure=f, width_ratios= [1,4,1])
for i in range(len(original_idxs)):
    # Get index of mode to plot
    j = original_idxs[i]
    # Two axes, which will contain maps of the phase and magnitude of a dynamic mode
    ax_map_1 = f.add_subplot(spec[i,0], projection=ccrs.AlbersEqualArea(central_longitude=-98.58, central_latitude=39.83))
    ax_map_2 = f.add_subplot(spec[i,2], projection=ccrs.AlbersEqualArea(central_longitude=-98.58, central_latitude=39.83))
    # Plot the phase and magntiude the given axes
    plot_vector_on_map(ax_map_1, np.angle(phi[:,j], deg=True))
    plot_vector_on_map(ax_map_2, np.abs(phi[:,j]), False)
    
    # Axis to plot the timeseries
    ax_tseries = f.add_subplot(spec[i,1])
    # Plot the timeseries
    ax_tseries.plot(gen_tseries(omega[j], t), color='#168c84')
    ax_tseries.set_ylabel("e^(omega t)", fontproperties=prop, fontsize=11)
    ax_tseries.set_xticks(date_idxs)
    ax_tseries.set_xticklabels([])
    yticks = ax_tseries.get_yticks()
    ax_tseries.set_yticklabels([round(q,2) for q in yticks], fontproperties=prop, fontsize=10)
    sns.despine(top=True, bottom=True, left=True, right=True)

# Only label the bottom timeseries
ax_tseries.set_xlabel("Time", fontproperties=prop, fontsize=12)
ax_tseries.set_xticks(date_idxs)
ax_tseries.set_xticklabels(plot_dates, fontproperties=prop, fontsize=10)
plt.savefig("../figures/chapter_2/dmd.svg", bbox_inches='tight')
plt.show()
```

```python
## Reconstruct the data - this also takes a minute or two to run

# Get the initial conditions
init_cond = np.linalg.lstsq(phi, X_prime[:,0], rcond=None)[0]

# Matrix to store the modes
u_modes = np.zeros((phi.shape[1], len(date_range)))
for i in range(len(date_range)):
    # Fill matrix 
    u_modes[:,i] = np.multiply(init_cond, np.exp(omega*i))
# The reconstruction. We only keep the real parts (the imaginary parts are basically zero, we just want to avoid them 
# and the real part indicate disease severity)
u_dmd = (phi@u_modes).real

# Select 6 days, evenly spaced throughout our date range, to compare the actual system to the DMD reconstructed system
ds = np.linspace(0, len(date_range)-2, 5)
# Convert to ints
ds = [int(q) for q in ds]

# Construct plot
f = plt.figure(figsize=(10,5))
f.subplots_adjust(hspace=0.01)
for i in range(5):
    # Date to plot
    j = ds[i]
    # Axis to store a map of the actual system
    ax_map_1 = f.add_subplot(2, 5, i+1, projection=ccrs.AlbersEqualArea(central_longitude=-98.58, central_latitude=39.83))
    # Axis to store a map of the DMD reconstructed system
    ax_map_2 = f.add_subplot(2, 5, i+6, projection=ccrs.AlbersEqualArea(central_longitude=-98.58, central_latitude=39.83))
    # Plot each vector
    plot_vector_on_map(ax_map_1, X_prime[:,j], rescale=False, colorbar=False)
    plot_vector_on_map(ax_map_2, u_dmd[:,j], rescale=False,  colorbar=True)
    
    # Labels for first row
#     if i==0:
#         ax_map_2.set_title("DMD reconstructed system \non %s"%(date_range[j+1]), fontproperties=prop, fontsize=12)
#         ax_map_1.set_title("Actual system \non %s"%(date_range[j+1]), fontproperties=prop, fontsize=12)
#     # Labels for every other row
#     else:
    ax_map_1.set_title((date_range[j+1]), fontproperties=prop, fontsize=12)
#         ax_map_2.set_title((date_range[j+1]), fontproperties=prop, fontsize=12)
# f.tight_layout()
plt.savefig("../figures/chapter_2/reconstruction.svg", dpi=200, bbox_inches='tight')
plt.show()
```

### Section 4: Prospective Analysis and Figures

```python
## Set up the data 

# We'll start our DMD using only 80% of days
start_t = int(0.8*n_days)

# We'll have to go back and remove counties that don't have any cases before our start date
zero_range = date_range[:start_t-1]
# List to store indices of counties with no deaths in our new date range
all_zero = []

for i in range(len(all_data)):
    tseries = df[zero_range].values[i]
    if np.all(tseries == 0):
        all_zero.append(i)

# The data we will use
pred_data = df.drop(df.index[all_zero], axis=0)
```

```python
## Filter for counties with more than 100,000 residents. Comment this line to remove this filter
pred_data = pred_data[pred_data.Population > 100000]
# Matrix of data
all_pred_data = pred_data[date_range].values
print("Shape of data", all_pred_data.shape)
```

```python
def calc_dmd(t, r=150):
    '''
    Perform dmd using up to t weeks and the first r singular values. Returns phi, the DMD modes,
    and omega, a diagonal matrix of the log of the eigenvalues.
    '''
    # Calculate z-scores
    X = scipy.stats.zscore(all_pred_data[:,:(t-1)], axis=1)
    X_prime = scipy.stats.zscore(all_pred_data[:,1:t], axis=1)

    # SVD X
    u,s,vh = np.linalg.svd(X, full_matrices=True)
    v = vh.T
    
    # Dimension reduction
    ur, sr, vr = u[:,:r], np.diag(s[:r]), v[:,:r]
    A_tilde = ur.T@X_prime@vr@np.linalg.inv(sr)
    
    # Perform DMD on the reconstructed data
    vals,vecs = np.linalg.eig(A_tilde)
    phi = X_prime@vr@np.linalg.inv(sr)@vecs
    omega = np.log(vals)
    return phi, omega
    
def predict_vals(t):
    '''
    Predict the state of the system 1 day after the given day, t.
    '''
    # Get phi and omega from DMD
    phi, omega = calc_dmd(t)
    # Get b using week t as our initial conditions
    b = np.linalg.lstsq(phi, all_pred_data[:,t], rcond=0)[0]
    # Project into the future by 1 day
    modes = np.multiply(b, np.exp(omega))
    return phi@modes

# Calculate prediction for all days, starting at day start_t. Store the results in res
res = np.zeros((all_pred_data.shape[0], n_days-start_t))
for i in range(start_t,n_days):
    p = predict_vals(i)
    res[:,i-start_t] = p
```

```python
## Get the actual z-scores to compare our predictions against

# Get the z-score of the original data
compared_data = scipy.stats.zscore(all_pred_data, axis=1)
# Filter to only the days we are predicting 
compared_data = compared_data[:,start_t:]
```

```python
# Calculate the MSE using predictions and actual z-score
mse = np.sum(np.sqrt(np.sum((compared_data - res)**2, axis=1))).real / len(res)
print("MSE: ", mse)
```
